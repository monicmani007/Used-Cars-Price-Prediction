```python
import pandas as pd
import numpy as np
import scipy
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import sklearn
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
```

```python
df["price"] = np.log1p(df["price"])
```

```python
df.head()
```

```python
df1 = pd.DataFrame()
```

### For Numerical Data

#### year

```python
year_min = df["year"].min()
year_max = df["year"].max()
df["year"] = ((df["year"] - year_min) / (year_max - year_min))
```

```python
df1['year'] = df["year"]
```

#### odometer

```python
odo_min = df["odometer"].min()
odo_max = df["odometer"].max()
df["odometer"] = ((df["odometer"] - odo_min) / (odo_max - odo_min))
```

```python
df1['odometer'] = df["odometer"]
```

### For Ordinal Data

#### cylinders

```python
label_encoder = LabelEncoder()
label_encoder.fit_transform(list(df['cylinders'].astype(str).values))
df['cylinders'] = label_encoder.transform(list(df['cylinders'].astype(str).values))
```

```python
df.head()
```

```python
#Normalizing Cylinder
cyl_min = df["cylinders"].min()
cyl_max = df["cylinders"].max()
df["cylinders"] = ((df["cylinders"] - cyl_min) / (cyl_max - cyl_min))
```

```python
df1['cylinders'] = df["cylinders"]
```

### For Categorical Data

```python
df1.reset_index(drop=True, inplace=True)
```

```python
cat = ['manufacturer', 'condition', 'fuel', 'title_status', 'transmission', 'type', 'state']
```

```python
for i in cat:
    encoder = OneHotEncoder(handle_unknown='ignore')
    label_encoder = LabelEncoder()
    temp = pd.DataFrame(encoder.fit_transform(df[[i]]).toarray())
    indexes = label_encoder.fit_transform(df[i])
    labels = label_encoder.inverse_transform(indexes)
    temp = temp.rename(columns={indexes[t]:str(i)+ str('_') + str(labels[t]) for t in range(len(labels))})
    df1 = pd.concat([df1, temp], axis=1)
```
